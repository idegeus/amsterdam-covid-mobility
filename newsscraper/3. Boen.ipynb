```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Load the weird excel-sheet.
sheet = pd.read_excel('./3. Maatregelen2020mbt covid.xlsx', names=np.arange(0, 24))

# Loop over each pair of two columns.
total_list = []
for month in np.arange(0, 12):
    
    # Month + day as both zero-padded string.
    month_padded = f"2020-{month + 1:02}-"
    days_series  = sheet[month*2].str[:-3].str.strip().str.pad(2, 'left', '0')
    days = ( month_padded + days_series).to_list()
    
    # Get the current messages
    msgs = sheet[month*2+1].to_list()
    
    # Zip together and add to the list. 
    total_list += list(zip(days, msgs))
    
total_list#[50:60]
```

```python
# Read list of items.
df = pd.DataFrame(total_list, columns=['date', 'measures'])
df = df.dropna(how='all')

df['date'] = pd.to_datetime(df.date, errors='raise', format='%Y-%m-%d')
df['measures'] = df.measures.astype(str).replace('nan', '')
df['weeknumber'] = df.date.dt.week
df['wk_monday'] = df.date - pd.to_timedelta(df.date.dt.weekday, unit='d')

df.to_csv('./data-maatregelen-boen.daylevel.csv')
df
```

```python
df_empty = df[df.measures != '']
df_empty = df_empty.groupby('wk_monday').measures.apply(lambda x: '\n'.join(x))
df_empty = pd.DataFrame(df_empty)

df_empty.to_csv('./data-maatregelen-boen.weeklevel.csv')
df_empty
```
